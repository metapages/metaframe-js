# Load a PDB File and Pass to Metaframe Visualization Widget

Read a local file in Python and push its content as inputs to a metaframe widget.

```python
from metaframe_widget import MetaframeWidget
```

## 1. Load the file

```python

with open("12AF.cif", 'r') as file:
    pdb_file = ''
    line = file.readline()
    
    while line:
        pdb_file += line
        line = file.readline()
```

## 2. Create a widget that renders the data as a table

```python
viewer = MetaframeWidget(url="https://js.mtfm.io/j/ddc0beae377df40ceafc9a9c1233340cabb3d79c86e15a71691baa765a5f0860", height="500px")

viewer
```

## 3. Pass the file content to the widget

```python
viewer.set_inputs({"cif": pdb_file})
```

## 4. Read the widget's outputs

```python
# Outputs arrive asynchronously — re-run this cell after a moment if it shows {}
viewer.outputs
```
