# Load a File and Pass to Metaframe Widget

Read a local file in Python and push its content as inputs to a metaframe widget.

```python
from metaframe_widget import MetaframeWidget
```

## 1. Load the file

```python
import csv

with open("data.csv") as f:
    reader = csv.DictReader(f)
    rows = list(reader)

print(f"Loaded {len(rows)} rows")
rows[:3]
```

## 2. Create a widget that renders the data as a table

```python
viewer = MetaframeWidget.from_code("""
export const onInputs = (inputs) => {
    const rows = inputs.rows;
    if (!rows || rows.length === 0) {
        document.getElementById("root").textContent = "No data";
        return;
    }
    const cols = Object.keys(rows[0]);
    const table = document.createElement("table");
    table.style.cssText = "border-collapse:collapse;width:100%;font-family:sans-serif;font-size:14px";

    const thead = table.createTHead();
    const hr = thead.insertRow();
    cols.forEach(col => {
        const th = document.createElement("th");
        th.textContent = col;
        th.style.cssText = "border:1px solid #ccc;padding:6px 10px;background:#f0f0f0;text-align:left";
        hr.appendChild(th);
    });

    const tbody = table.createTBody();
    rows.forEach(row => {
        const tr = tbody.insertRow();
        cols.forEach(col => {
            const td = tr.insertCell();
            td.textContent = row[col];
            td.style.cssText = "border:1px solid #ccc;padding:6px 10px";
        });
    });

    const root = document.getElementById("root");
    root.innerHTML = "";
    root.appendChild(table);

    setOutput("rows", rows);
    setOutput("count", rows.length);
};
""", height="300px")

viewer
```

## 3. Pass the file content to the widget

```python
viewer.set_inputs({"rows": rows})
```
