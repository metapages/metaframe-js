# Metaframe Widget Demo

Use any metaframe as a Jupyter widget. Just paste a URL.

```python
from metaframe_widget import MetaframeWidget
```

## 1. Create a widget from a URL

```python
w = MetaframeWidget(url="https://framejs.io/")
w
```

## 2. Create from inline code

```python
echo = MetaframeWidget.from_code("""
export const onInputs = (inputs) => {
    document.getElementById("root").textContent = JSON.stringify(inputs);
    // Pass each input key through as a separate output
    Object.keys(inputs).forEach(key => setOutput(key, inputs[key]));
};
""")
echo
```

## 3. Push inputs from Python

```python
echo.set_inputs({"data": [1, 2, 3], "message": "hello from Python"})
```

## 4. Read outputs

```python
# Outputs arrive asynchronously — re-run this cell after a moment if it shows {}
echo.outputs
```

## 5. React to output changes

```python
echo.on_outputs_change(lambda change: print("Got:", change["new"]))
```

## 6. Pipe widgets together

```python
doubler = MetaframeWidget.from_code("""
export const onInputs = (inputs) => {
    if (inputs.data) {
        const doubled = inputs.data.map(x => x * 2);
        document.getElementById("root").textContent = JSON.stringify(doubled);
        setOutput("doubled", doubled);
    }
};
""")

# Connect echo's "data" output to doubler's "data" input
echo.pipe_to(doubler, output_key="data", input_key="data")

doubler
```

```python
# Push data through the pipeline
echo.set_inputs({"data": [10, 20, 30]})
```

```python
# Outputs arrive asynchronously — re-run this cell after a moment if it shows {}
doubler.outputs
```
